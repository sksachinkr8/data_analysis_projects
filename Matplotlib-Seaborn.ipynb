## MATPLOTLIB

This library is core for creating interactive visualization in Python.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
Year = [1920, 1930, 1940, 1950, 1960, 1970, 1980, 1990, 2000, 2010] 

Unemployment_Rate = [9.8, 12, 8, 7.2, 6.9, 7, 6.5, 6.2, 5.5, 6.3] 

Unemployment_Rate1 = [1.8, 2,8, 4.2, 6.0, 7,3.5, 5.2, 7.5, 5.3] 

plt.plot(Year, Unemployment_Rate)
```

By default, it is plotting a line plot.

We can add title to the plot and labels to the axes.

```python
plt.plot(Year, Unemployment_Rate)
plt.title('Unemployment Rate versus Year')
plt.xlabel('Year')
plt.ylabel('Unemployment Rate')
plt.show()
```

We can show the grid lines on the plot as well.

```python
plt.plot(Year, Unemployment_Rate)
plt.title('Unemployment Rate versus Year')
plt.xlabel('Year')
plt.ylabel('Unemployment Rate')
plt.grid()
plt.show()
```

Let's plot both Unemployment_Rate and Unemployment_Rate1 on the same plot.

```python
plt.plot(Year, Unemployment_Rate)
plt.plot(Year, Unemployment_Rate1)
plt.title('Unemployment Rate versus Year')
plt.xlabel('Year')
plt.ylabel('Unemployment Rate')
plt.grid()
plt.show()
```

In order to distinguish two data sets in one plot, we need to set labels in plt.plot and then set the labels as legends.

```python
plt.plot(Year, Unemployment_Rate, label = "UR1")
plt.plot(Year, Unemployment_Rate1, label = "UR2")
plt.title('Unemployment Rate versus Year')
plt.xlabel('Year')
plt.ylabel('Unemployment Rate')
plt.legend()
plt.show()
```

We can also put markers on the plot to idenitfy them better. We can also pass the size of the marker as an argument.

```python
plt.plot(Year, Unemployment_Rate, marker = '*', markersize = 10, label = "UR1")
plt.plot(Year, Unemployment_Rate1, marker = '>', markersize = 10, label = "UR2")
plt.title('Unemployment Rate versus Year', fontsize = 14)
plt.xlabel('Year')
plt.ylabel('Unemployment Rate', fontsize = 12)
plt.legend()
plt.show()
```

## Bar Graphs

So far, we have only discussed how to create a bar plot.

Say, we want to create a bar graph or plot.

```python
Country = ['USA', 'Canada', 'Germany', 'UK', 'France'] 

GDP_Per_Capita = [45000, 42000, 52000, 49000, 47000] 
 
plt.bar(Country, GDP_Per_Capita)
```

Just like before we can add other attributes to the plot, such as colors, title, axis labels, legends, etc.

```python
col = ['#FFD700', '#ADFF2F', '#008000', '#00FF7F']

plt.bar(Country, GDP_Per_Capita, color = col) 
plt.title("Country vs GDP Per Capita") 
plt.xlabel("Country") 
plt.ylabel("GDP Per Capita")
```

## Subplots

The subplots() function of Matplotlib is used to create a figure and a set of subplots, i.e., we can create multiple plots in the plotting area using subplots() function.

```python
fig, ax = plt.subplots(1,2)

ax
```

The parameters in the subplot function are the number of rows and columns.

'ax' is like an array of subplots.

```python
ax.shape
```

Let's plot the precious plots in these subplots.

```python
ax[0].bar(Country, GDP_Per_Capita, color = col)
ax[1].plot(Year, Unemployment_Rate, marker = '*', markersize = 10)
fig
```

We can vertically stack these plots as well.

```python
fig, ax = plt.subplots(2,1)

ax[0].bar(Country, GDP_Per_Capita, color = col)
ax[1].plot(Year, Unemployment_Rate, marker = '*', markersize = 10)
```

Let's add some attributes to the subplots.

```python
fig, ax = plt.subplots(2,1)

ax[0].bar(Country, GDP_Per_Capita, color = col)
ax[1].plot(Year, Unemployment_Rate, marker = '*', markersize = 10)
ax[0].set_title("GDP Per Capita vs Country")
ax[1].set_title("Unemployment Rate vs Year")
ax[0].set_ylabel("GDP Per Capita")
ax[1].set_ylabel("Unemployment Rate")
fig.subplots_adjust(hspace=0.4)
```

## Scatter Plots

Let's look at the Scatter Plots.

```python
l1 = np.random.randint(1, 10, size = 10)
l2 = np.random.randint(100, 200, size = 10)

plt.scatter(l1, l2)
```

```python
from pydataset import data

mt = data('mtcars')
mt.head()
```

```python
plt.scatter(mt['mpg'], mt['disp'])
```

From this graph, we can understand if there is a relationship between the data sets. If so, we can use a linear regression or a clustering algorithm.

Let's say we want to display 3 variables in a scatter plot. Usually the scatter plot is meant only for two variables. However, we can exploit the size to display 3rd variable.

```python
plt.scatter(mt['mpg'], mt['disp'], s = 100)
```

```python
plt.scatter(mt['mpg'], mt['disp'], s = mt['wt']*100)
```

We can use color to represent the fourth dimension.

```python
plt.scatter(mt['mpg'], mt['disp'], c = mt['wt'], s = mt['cyl']*100)
```
