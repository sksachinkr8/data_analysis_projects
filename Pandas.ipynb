## Table Concatenation

```python
grades1 = {'subject1': ['A1', 'B1', 'A2', 'A3'], 'subject2': ['A2', 'A1', 'B2', 'B3']}
df1 = pd.DataFrame(grades1)
df1
```

```python
grades2 = {'subject3': ['A1', 'B1', 'A2', 'A3'], 'subject4': ['A2', 'A1', 'B2', 'B3']}
df2 = pd.DataFrame(grades2)
df2
```

```python
pd.concat([df1, df2])
```

By default, columns are getting combined row-by-row. But since keys are different, the data is combined as shown above. 

```python
grades3 = {'subject2': ['A1', 'B1', 'A2', 'A3'], 'subject4': ['A2', 'A1', 'B2', 'B3']}
df3 = pd.DataFrame(grades3)
df3
```

```python
pd.concat([df1, df3])
```

If the keys are same the rows are concatenated in the same column as shown above.
By default, we are working with axis = 0. If we change the axis to 1, the concatenation will work in the index of the rows.

```python
pd.concat([df1, df3], axis = 1)
```

```python
pd.concat([df1, df3], axis = 0, ignore_index = True)
```

If we use ignore_index = True, it will reset the index of the concatenated table.

## Merging Data

Four types of Merge:
1. Inner Join
2. Left Join
3. Right Join
4. Full Outer Join

In order to merge, we require a common key in the two tables.

```python
rollno = pd.Series(range(1, 11))
rollno 
```

```python
l1 = ["Student"+str(i) for i in range(1,11)]
l1
```

```python
genderlist = ["M", "F"]
gender = np.random.choice(genderlist, size = 10)
gender
```

```python
marks1 = np.random.randint(0, 100, size = 10)
marks1
```

```python
pd1 = pd.DataFrame({'rollno': rollno, 'name': l1, 'gender': gender, 'marks': marks1})
pd1
```

```python
course = np.random.choice(a = ['BBA', 'MBA', 'BTech'], size = 10)
course
```

```python
marks2 = np.random.randint(40, 100, size = 10)
marks2
```

```python
pd2 = pd.DataFrame({'rollno': rollno, 'course': course})
pd2
```

```python
fees = pd.DataFrame({'course': ['BBA', 'MBA', 'BTech', 'MTech'], 'fees': [100000, 200000, 150000, 220000]})
fees
```

Let's merge the dataframes pd1 and pd2

```python
pd3 = pd.merge(pd1, pd2)
pd3
```

```python
rollno1 = pd.Series(range(6, 16))
rollno1
```

Let's create another table with roll number for 6 to 15

```python
pd4 = pd.DataFrame({'rollno': rollno1, 'course': course})
pd4
```

```python
# Inner Join
pd5 = pd.merge(pd1, pd4, how = 'inner')
pd5
```

```python
# Outer Join
pd6 = pd.merge(pd1, pd4, how = 'outer')
pd6
```

```python
# Left Join
pd7 = pd.merge(pd1, pd4, how = 'left')
pd7
```

```python
# Right Join
pd8 = pd.merge(pd1, pd4, how = 'right')
pd8
```

Now, suppose the name of the commom key is not same in two tables. In this case, the merge will show an error that there are no common columns for the merger. However, we can provide the columns on which merge is to take place.

```python
pd4 = pd.DataFrame({'rollno1': rollno1, 'course': course})
pd4
```

```python
pd9 = pd.merge(pd1, pd4, how = 'inner', left_on = 'rollno', right_on = 'rollno1')
pd9
```

## Grouping Data

Grouping allows us to group data and apply functions on these groups.

```python
pd4.groupby('course').size()
```

```python
pd4.groupby('course').count()
```

We can have multiple labels to group the data. We can pass a list of labels to the groupby function.

```python
pd9.groupby(['course', 'gender']).size()
```

```python
pd9.groupby(['course', 'gender']).agg({'marks': ["mean", "std", "max", "min"]})
```
